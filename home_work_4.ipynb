```python
# 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
# Найдите ее среднее значение и дисперсию.
# Среднее значение: M(X) = (a + b) / 2
M1 = (200 + 800) / 2
M1
```

```python
# Дисперсия: D(X) = (b - a) ** 2 / 12
D1 = (800 - 200) ** 2 / 12
D1
```

```python
print(F'Ответ: Среднее значение = {M1}, дисперсия = {D1}')
```

```python
# 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
# Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
# Если да, найдите ее.
# (b - 0.5) ** 2 / 12 = 0.2
# (b ** 2 - 2 * b * 0.5 + 0.25) / 12 = 0.2
# b ** 2 - b + 0.25 = 0.2 * 12
# b ** 2 - b = 2.4 - 0.25
#  Получилось квадратное уравнение
# b ** 2 - b - 2.15 = 0
# Решение квадратного уравнения
a =  1
b = -1
c = -2.15
# Надо найти дискриминант
dis = b ** 2 - 4 * a * c
print('Дискриминант квадратного уравнения =',dis)
# dis = 1 + 4 * 1 * 2.15
# dis = 9.6
# Нахожу кортни квадратного уравнения
from math import sqrt
x1 = (-b + sqrt(dis)) / (2 * a)
print('Первый корень =', x1)
x2 = (-b - sqrt(dis)) / (2 * a)
print('Второй корень =', x2)
print(F'Правая граница ({x2}) не может быть < левой (0.5), значит правая граница = {x1}')
```

```python
print(F'Ответ: правая граница величины B = {x1}, ее среднее значение = {M2}')
```

```python
# 4. Рост взрослого населения города X имеет нормальное распределение.
# Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
# Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
# а). больше 182 см
Z1 = (182 - 174) / 8 
print(F'Найду в таблице накопленного нормального распределения {Z1}')
P1 = 1 - 0.8413
print(F'Ответ: вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 182 см, = {P1}')
```
